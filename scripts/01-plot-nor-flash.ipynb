```python
import numpy as np
import matplotlib.pyplot as plt

%matplotlib notebook
```

```python
bs = open("../roms/nor_flash_first_8MB.bin", "rb").read()
```

```python
xs = np.unpackbits(np.frombuffer(bs, dtype=np.uint8))

xs_pop_count = xs.reshape((-1, 16)).sum(axis=1)
```

```python
fig = plt.figure()

ax = fig.subplots()
ax.plot(xs_pop_count)

ax.set_title("Content of the NOR Flash (S29GL01GS) on the Violin Stratix V board")
ax.set_ylabel("pop count / block")
ax.set_xlabel("block index")

ax.grid(True)
```
